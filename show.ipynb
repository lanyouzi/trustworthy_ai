```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
basedir = os.getcwd()
ds_basedir = os.path.abspath("E:\datasets")
if not os.path.isdir('results'):
    os.mkdir('results')
out_path = os.path.join(basedir, 'results')
```

```python
train_acc_history_lr_1e_4 = np.load('./results/train_acc_history_lr_0.0001.npy')
train_loss_history_lr_1e_4 = np.load('./results/train_loss_history_lr_0.0001.npy')
val_acc_history_lr_1e_4 = np.load('./results/val_acc_history_lr_0.0001.npy')
val_loss_history_lr_1e_4 = np.load('./results/val_loss_history_lr_0.0001.npy')
print('acc:', train_acc_history_lr_1e_4.max(), val_acc_history_lr_1e_4.max())
print('loss:', train_loss_history_lr_1e_4.min(), val_loss_history_lr_1e_4.min())

plt.figure()
plt.plot(range(len(train_loss_history_lr_1e_4)), train_loss_history_lr_1e_4, label='train_loss')
plt.plot(range(len(val_loss_history_lr_1e_4)), val_loss_history_lr_1e_4, label='val_loss')
plt.title('loss trend when lr=1e-4')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend()
plt.savefig(os.path.join(out_path, 'loss_lr_0.0001.png'))

plt.figure()
plt.plot(range(len(train_acc_history_lr_1e_4)), train_acc_history_lr_1e_4, label='train_acc')
plt.plot(range(len(val_acc_history_lr_1e_4)), val_acc_history_lr_1e_4, label='val_acc')
plt.title('acc trend when lr=1e-4')
plt.ylabel('acc')
plt.xlabel('epoch')
plt.legend()
plt.savefig(os.path.join(out_path, 'acc_lr_0.0001.png'))
```
